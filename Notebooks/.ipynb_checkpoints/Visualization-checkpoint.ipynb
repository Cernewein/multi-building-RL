```python
zeta0_5_above = 402.0
zeta0_2_above = 247.0
data_base_above = 494.4

labels_values = np.round(np.array([zeta0_5_above,zeta0_2_above, data_base_above]), decimals=0)
costs = labels_values
labels = ['Zeta 0.5', 'Zeta 0.2', 'Base']


N=3
ind = np.arange(N)
f, ax = plt.subplots(figsize=(15,7))
ax.bar(ind, costs, bottom=25)
#moving_avg.plot(ax=ax)
ax.set_ylabel('Exceeding load (kWh)')
ax.set_yticks(np.arange(0, 500, 100))
#ax.set_title('Cost incurred by the different solutions')
for i, v in enumerate(labels_values):
    ax.text(i-0.1, 
              v+30, 
              v, 
              fontsize=12)
ax.set_xticks(ind)
ax.set_xticklabels(labels)
plt.savefig('data/output/BaseProblem/comparing_different_zetas_to_base.png',bbox_inches='tight')
```
